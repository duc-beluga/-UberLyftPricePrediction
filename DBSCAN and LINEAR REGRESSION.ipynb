```python
import numpy as np
from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
from pandas import DataFrame
import pandas as pd
from sklearn.cluster import DBSCAN
```

```python
df = pd.read_excel('Lyft.xlsx')

df = df.iloc[:-1, :]
df = df.dropna()
df_np = df.to_numpy()
dist = df_np[:, 0]
price = df_np[:, 1]
plt.subplots(figsize=(8,8))
plt.scatter(x=dist, y=price)
plt.show()

clustering = DBSCAN(eps=2.5, min_samples=6).fit(df_np)
cluster = clustering.labels_
print(len(set(cluster)))
df = DataFrame(dict(x=dist, y=price, label=cluster))
df_grouped = df.groupby('label')
fig, ax = plt.subplots(figsize=(8,8))
colors = {-1: 'red', 0: 'blue', 1: 'orange', 2: 'green', 3: 'skyblue', 4: 'cyan', 5: 'yellow', 6: 'black', 7: 'magenta'}
for key, group in df_grouped:
    group.plot(ax=ax, kind="scatter", x='x', y='y', label=key, color=colors[key])
plt.xlabel('Distance')
plt.ylabel('Price')
plt.show()
```

```python
df_filtered = df[df.label>-1]
df_grouped = df_filtered.groupby('label')
fig, ax = plt.subplots(figsize=(8,8))
for key, group in df_grouped:
    group.plot(ax=ax, kind="scatter", x='x', y='y', label=key, color=colors[key])
plt.xlabel('Distance')
plt.ylabel('Price')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_excel('Lyft.xlsx')

df = df.iloc[:-1, :]
df = df.dropna()
df_np = df.to_numpy()
dist = df_np[:, 0]
dist = dist[:, np.newaxis]
price = df_np[:, 1]
price = price[:, np.newaxis]
plt.subplots(figsize=(8,8))
plt.scatter(x=dist, y=price)
plt.show()
regression = LinearRegression().fit(dist, price)
x_test = np.linspace(-3,3)
y_pred = regression.predict(x_test[:, None])

plt.scatter(dist,price)
plt.plot(x_test, y_pred, 'r')
plt.legend(['Predicted Line', 'Observed Data'])
plt.show()
```
